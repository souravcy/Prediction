```python
import os
import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import tensorflow as tf
```

```python
data=pd.read_csv("data/city_hour.csv")
```

```python
data=data[data["City"]=="Kolkata"]
```

```python
data.drop(["City"],axis=1,inplace=True)
```

```python
data=data.drop('AQI_Bucket',axis=1)
data
```

```python
df=data
df=df.reset_index().drop('index',axis=1)
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3)
pf=df
df=df.drop('Datetime',axis=1)
imputed = imputer.fit_transform(df)
df = pd.DataFrame(imputed, columns=df.columns)
```

```python
df
```

```python
df['Date']=pf['Date']
df = df[df['Date'] < ('2020-01-01')]
df
```

```python
def df_to_X_y(df, window_size=5):
  df_as_np = df.to_numpy()
  X = []
  y = []
  for i in range(len(df_as_np)-window_size):
    row = [[a] for a in df_as_np[i:i+window_size]]
    X.append(row)
    label = df_as_np[i+window_size]
    y.append(label)
  return np.array(X), np.array(y)
```

```python
temp=df['AQI']
temp.plot()
temp
```

```python
WINDOW_SIZE = 5
X1, y1 = df_to_X_y(temp, WINDOW_SIZE)
X1.shape, y1.shape
print(X1[0])
print(y1[0])
```

```python
temp=df['AQI']
temp.plot()
temp
```

```python
X_train1, y_train1 = X1[:400], y1[:400]
X_val1, y_val1 = X1[400:500], y1[400:500]
X_test1, y_test1 = X1[500:], y1[500:]
X_train1.shape, y_train1.shape, X_val1.shape, y_val1.shape, X_test1.shape, y_test1.shape
```

```python
WINDOW_SIZE = 5
X1, y1 = df_to_X_y(temp, WINDOW_SIZE)
X1.shape, y1.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

model1 = Sequential()
model1.add(InputLayer((5, 1)))
model1.add(LSTM(64))
model1.add(Dense(8, 'relu'))
model1.add(Dense(1, 'linear'))

model1.summary()
model=model1
```

```python
#cp1 = ModelCheckpoint('model1/', save_best_only=True)
model.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.0001), metrics=[RootMeanSquaredError()])
```

```python
model.fit(X_train1, y_train1, validation_data=(X_val1, y_val1), epochs=100)
```

```python
from tensorflow.keras.models import load_model
model1 = load_model('model1/')
```
